```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('50_Startups.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python

X = df.iloc[:,0:3]
Y = df.iloc[:,4:5]

X.head()

```

```python
Y.head()
```

```python
from sklearn.impute import SimpleImputer

si = SimpleImputer()
X.iloc[:,0:3] = si.fit_transform(X.iloc[:,0:3])
X
```

```python
plt.scatter(X.iloc[:,0:1],Y)
```

```python
plt.scatter(X.iloc[:,1:2],Y)
```

```python
plt.scatter(X.iloc[:,2:3],Y)
```

```python
print(X[:10])
```

```python
X[:10]
```

```python

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
# normalization -> (x - mean) / std -> unnomalization -> (x*std) + mean
X = sc.fit_transform(X)
X
```

```python
from sklearn.linear_model import LinearRegression

mlr = LinearRegression()
```

```python
mlr.fit(X, Y)
```

```python
mlr.score(X, Y) * 100
```

```python
mlr.coef_ , mlr.intercept_
```

```python
import pickle
filename = 'linear_model.sav'
pickle.dump(mlr, open(filename, 'wb')) 
pickle.dump(sc, open('scaler.sav', 'wb'))
```
